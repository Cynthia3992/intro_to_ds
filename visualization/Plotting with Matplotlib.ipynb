# Data Visualization using Matplotlib

Matplotlib is a 2-D plotting library that helps in visualizing figures. Matplotlib emulates Matlab like graphs and visualizations. Matlab is not free, is difficult to scale and as a programming language is tedious. So, matplotlib in Python is used as it is a robust, free and easy library for data visualization.

#### Intro to pyplot

matplotlib.pyplot is a collection of functions that make matplotlib work like MATLAB. Each pyplot function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc.

In matplotlib.pyplot various states are preserved across function calls, so that it keeps track of things like the current figure and plotting area, and the plotting functions are directed to the current axes (please note that "axes" here and in most places in the documentation refers to the axes part of a figure and not the strict mathematical term for more than one axis).


Refer to the [matlotlib documenation]() for more information on the concepts below and more detailed documentation of the functions disucssed.

### Lets start with the basics

It may seems like there are lots of compoents and it may get overwhelming. Luckily, this library is  flexible and has a lot of handy, built-in defaults that will help you out tremendously. As such, you donâ€™t need much to get started: you need to make 

* `plt.plot()` for plotting line chart similarly in place of plot other functions are used for plotting. All plotting functions require data and it is provided in the function through * parameters. 
* `plt.legend()` for signifying the observation variables.  
* `plot.show()` for displaying the plot  


```python
# Prepare the data
x = np.linspace(0, 10, 100)

# Plot the data
plt.plot(x, x, label='linear')

# Add a legend
plt.legend()

# Add a title
plt.title('My First Plot')

# Show the plot
plt.show()

```


That was easy! 

You are actually using more functions than what you see here!

Theere are two big components that you need to take into account:


The Figure is the overall window or page that everything is drawn on. Itâ€™s the top-level component of all the ones that you will consider in the following points. You can create multiple independent Figures. A Figure can have several other things in it, such as a suptitle, which is a centered title to the figure. Youâ€™ll also find that you can add a legend and color bar, for example, to your Figure. It is crucial as it is the object that actually does the drawing to get you your plot, but as the user it is more-or-less invisible to you). 


To the figure you add Axes. The Axes is the area on which the data is plotted with functions such as plot() and scatter() and that can have ticks, labels, etc. associated with it. This explains why Figures can contain multiple Axes. A figure can contain any number of Axes, but will typically have at least one



```python
# create a figure 
fig = plt.figure() # creates an empty figure with no Axes

# add an axes to the figure 
ax = fig.add_subplot() 

ax.set(xlim=[0.5, 4.5], ylim=[-2, 8], title='An Example Axes',
       ylabel='Y-Axis', xlabel='X-Axis')

plt.show()
```

```python
## notes: plt.subsubplots() and plt.subplot() are 2 different functions, with different inputs and outputs. 

fig, axes = plt.subplots(2, 2) # a figure with a 2x2 grid of Axes (nrows, ncols)
ax11, ax12, ax21, ax22 = axes.ravel() # Return a contiguous flattened array.

ax11.set_title("ax11")
ax12.set_title("ax12")
ax21.set_title("ax21")
ax22.set_title("ax22")
plt.tight_layout()

```

```python
fig = plt.figure()
ax13 = plt.subplot(3, 1, 1)
ax323 = plt.subplot(3, 2, 3)
ax324= plt.subplot(3, 2, 4)
ax33 = plt.subplot(3, 1, 3)


ax13.set_title("wide plot")
ax323.set_title("short plot 1")
ax324.set_title("short plot 2")
ax33.set_title("wide plot 2")

plt.tight_layout()
plt.show()


```

### Inputs

All of plotting functions expect numpy.array or numpy.ma.masked_array as input. Classes that are 'array-like' such as pandas data objects and numpy.matrix may or may not work as intended. It is best to convert these to numpy.array objects prior to plotting.



```python
x = np.linspace(0, 2, 100)

# Create a figure and an axes.
# figsize lets you define the size of figure: (width, height), in inches

fig, ax = plt.subplots(figsize = (10,8))  

ax.plot(x, x, label='linear')  # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.

ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.

ax.set_title("Simple Plot")  # Add a title to the axes.

ax.legend()  # Add a legend.
```

```python
# plt.figasepect()  defines the aspect ratio (i.e. the ratio height / width)

fig = plt.figure(figsize=plt.figaspect(.25))


linear = plt.subplot(1, 3, 1)
quadratic = plt.subplot(1, 3, 2)
cubic = plt.subplot(1, 3, 3)


linear.plot(x, x, color = 'forestgreen')
quadratic.plot(x, x**2, color = 'steelblue')
cubic.plot(x, x**3, color = 'lightcoral')

linear.set_title('linear')
quadratic.set_title('quadratic')
cubic.set_title('cubic')

plt.show()
```

## Histogram

A histogram takes in a series of data and divides the data into a number of bins. It then plots the frequency data points in each bin (i.e. the interval of points). It is useful in understanding the count of data ranges.

When to use: We should use histogram when we need the count of the variable in a plot.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html

inputs: 
x : array or squence   
bins: 
* If bins is an integer, it defines the number of equal-width bins in the range.

* If bins is a sequence, it defines the bin edges, including the left edge of the first bin and the right edge of the last bin; in this case, bins may be unequally spaced. All but the last (righthand-most) bin is half-open. In other words, if bins is:

```python
## histogram of the # of people visiting the GrandfCanyon per year
fig, ax = plt.subplots(2, 3, figsize=(20, 8))

ax[0,0].hist(np_data['GrandCanyon'], edgecolor = 'black')
ax[0,0].set_title('Default Image')

ax[0,1].hist(np_data['GrandCanyon'], edgecolor = 'black', bins = 5)
ax[0,1].set_title('5 Bins')

ax[0,2].hist(np_data['GrandCanyon'], edgecolor = 'black', bins = [1e6,6e6,8e6])
ax[0,2].set_title('Defined Bin Edges')

ax[1,0].hist(np_data['GrandCanyon'], edgecolor = 'black', cumulative = True)
ax[1,0].set_title('Cumulative')

ax[1,1].hist(np_data['GrandCanyon'], edgecolor = 'black', range = (2e6, 3.5e6))
ax[1,1].set_title('Range: 2M - 3M')


ax[1,2].hist(np_data['GrandCanyon'],  color = 'orange')
ax[1,2].hist(np_data['Badlands'], edgecolor = 'black')
ax[1,2].set_title('Multiple Histograms')



plt.tight_layout()
plt.show()
```

## Scatter Plot

Scatter plot helps in visualizing 2 numeric variables. It helps in identifying the relationship of the data with each variable i.e correlation or trend patterns. It also helps in detecting outliers in the plot.

When to use: It is used in Machine learning concepts like regression, where x and y are continuous variables. It is also used in clustering scatters or outlier detection.

```python
countries=pd.read_csv('datasets/countries.csv')
countries.head()
```

```python
data_2007=countries[countries['year']==2007]

plt.scatter(data_2007['gdpPerCapita'],data_2007['lifeExpectancy'])
plt.title('Life Expectancy vs GDP Per Capita in 2007 ')
plt.xlabel('GDP per Capita($)')
plt.ylabel('Life Expectancy')
plt.show()
```

```python
plt.scatter(np.log10(data_2007['gdpPerCapita']),data_2007['lifeExpectancy'])
plt.title('Life Expectancy vs GDP Per Capita in 2007 ')
plt.xlabel('GDP per Capita($)')
plt.ylabel('Life Expectancy')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(countries['year'],countries['gdpPerCapita'],countries['lifeExpectancy'], s=30)
plt.show()
```
